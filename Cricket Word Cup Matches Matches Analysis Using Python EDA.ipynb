```python
# Question = Display win margin between two teams
```

```python
result_margin=batting_frist["Score A"] - batting_frist["Score B"]
result_margin
```

```python
batting_frist["Wining Team"]
```

```python
plt.figure(figsize=(11,7))
plt.bar(batting_frist["Wining Team"],result_margin,color=["orange"])
plt.title("Winning Margin between two teams")
plt.show()
```

```python
plt.bar(batting_frist["Wining Team"][0:7],batting_frist["Score A"][0:7],color=["orange"])
plt.xlabel("Winning Team")
plt.ylabel("Score A")
plt.title("Top 7 Teams' Scores when Batting First")
plt.show()
```
